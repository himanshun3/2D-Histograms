```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('character-deaths.csv')
df.fillna(10,inplace=True)              #Replace all NaN with 10
x=df['Death Chapter']
y=df['Book Intro Chapter']
plt.hist2d(x,y,bins=(10,20))            #Rectangular bins
plt.colorbar()
plt.xlabel('Death Chapter')
plt.ylabel('Book Intro Chapter')
plt.show()
```
